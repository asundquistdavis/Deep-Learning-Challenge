# Part 3: Optimazation
The model created in AlphabetSoupCharity_DeepLearning.ipynb obtained an accuracy of 72.98%. In this notebook we'll try to improve this score by adjusting parameters such as the number of hidden layers, number of nodes per layer and the type of activation function.

```python
# import dependencies
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler as SS
import pandas as pd
import tensorflow as tf
from tensorflow import keras as k
import keras_tuner as kt
```

```python
# load and prepare the data for analysis (see ASC_DeepLearning for more info on these steps)
import pandas as pd 
df = pd.read_csv("Resources/charity_data.csv")
df = df.drop(['EIN', 'NAME'], axis=1)
df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].apply(lambda x: x if x in ['T3', 'T4', 'T6', 'T5', 'T19', 'T8', 'T7'] else 'other')
df['CLASSIFICATION'] = df['CLASSIFICATION'].apply(lambda x: x if x in ['C1000', 'C2000', 'C1200', 'C3000', 'C2100', 'C7000', 'C1700', 'C4000', 'C5000', 'C1270', 'C2700'] else 'other')
dfr = pd.get_dummies(df)
X = dfr.drop('IS_SUCCESSFUL', axis=1)
y = dfr['IS_SUCCESSFUL']
_X, X_, _y, y_ = tts(X, y)
scaler = SS()
scaler.fit(_X)
_Xs = scaler.transform(_X)
Xs_ = scaler.transform(X_)
```

## Attempt 1: Add More Layers

```python
model = k.models.Sequential()

model.add(k.layers.Dense(units=50, activation='relu', input_dim=48))

model.add(k.layers.Dense(units=50, activation='relu'))

model.add(k.layers.Dense(units=50, activation='relu'))

model.add(k.layers.Dense(units=50, activation='relu'))

model.add(k.layers.Dense(units=1, activation='sigmoid'))

model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

model.summary()
```

```python
model.fit(_Xs, _y, epochs=100)
```

```python
model_loss, model_accuracy = model.evaluate(Xs_, y_, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Attempt 4: Keras Tuner
Keras tuner is a tensor flow/keras module that "searches" for an optimal keras model. The tuner trains many models using testing different paramenters such as number of nodes, number of layers and types of sctivation functions. Since the number of layers did not change perfomrance much (see attempt 1) we will not have the tuner look at more than 3 layers hidden layers. Both the number of nodes and type of activation function seemed to significantly increase accuracy so we will have the tuner varry both of those. 

```python
def create_model(hp):
    
    # create the model
    model_4 = k.models.Sequential()

    # allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','elu'])
    
    # allow kerastuner to decide number of neurons in first layer
    model_4.add(k.layers.Dense(units=hp.Int('units_0', min_value=50, max_value=150, step=50), activation=activation, input_dim=48))

    # allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', min_value=1, max_value=3, step=1)-1):
        model_4.add(k.layers.Dense(units=hp.Int('units_' + str(i), min_value=50, max_value=150, step=50), activation=activation))
    
    model_4.add(k.layers.Dense(units=1, activation="sigmoid"))

    # compile the model
    model_4.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
    
    return model_4
```

```python
tuner = kt.Hyperband(create_model, objective="val_accuracy", overwrite=True, max_epochs=10, hyperband_iterations=2)

tuner.search(_Xs, _y, epochs=20, validation_data=(Xs_, y_))
```

```python
model_loss, model_accuracy  = tuner.get_best_models(1)[0].evaluate(Xs_, y_, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
